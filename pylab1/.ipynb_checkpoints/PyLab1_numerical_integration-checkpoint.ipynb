```python
#Import required packages
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Import the data 
time,distance,dist_err = np.loadtxt('pylab1_motiondata.txt', skiprows=2,unpack=True)

#Plot the data for a first look
plt.figure(figsize=(10,8))
plt.plot(time, distance, color='k')
plt.xlabel("Time (s)", fontsize=20)
plt.ylabel("Distance (cm)", fontsize=20)
plt.title('Plot of the Mass on a Spring Data', fontsize=25)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.grid()
#plt.savefig("dataplot.png")
```

```python
#Compute position and velocity using the Forward Euler method

#Define constants
dt = 0.01              #Timestep Delta t in seconds
k = 14.6               #Spring constant in N/m
m = 0.2                #Mass of bob in kg
Omega_0 = np.sqrt(k/m) #Angular frequency formula
t0 = 0                 #Initial time, in seconds
T = 0.735              #Period taken from motion sensor data, in seconds

ti = np.arange(0,10.0,dt) #Array of time values, in seconds

#Zero arrays for holding numerical solutions
y = np.zeros(len(ti)) #Array for position
v = np.zeros(len(ti)) #Array for velocity

#Initial conditions
y[0] = 0.021875 #Amplitude of experiment in m
v[0] = 0.0      #Initial velocity before release in m/s

#Numerical integration to solve for the position and velocity of the mass
for i in range(0,len(ti)-1): 
    y[i + 1] = y[i] + (dt*v[i]) 
    v[i + 1] = v[i] - (dt*(Omega_0**2)*y[i]) 
    
#Position vs. time plot
plt.figure(figsize=(10,8))
plt.plot(ti, y, color='k')
plt.xlabel("Time (s)", fontsize=20)
plt.ylabel("Distance (m)", fontsize=20)
plt.title("Position vs. Time for the Spring-Mass System (Forward Euler Method)", fontsize=25)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.grid()
#plt.savefig("pylab1pt1_pos_time_plot.png")

#Velocity vs. time plot
plt.figure(figsize=(10,8))
plt.plot(ti, v, color='k')
plt.xlabel("Time (s)", fontsize=20)
plt.ylabel("Velocity (m/s)", fontsize=20)
plt.title("Velocity vs. Time for the Spring-Mass System (Forward Euler Method)", fontsize=25)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.grid()
#plt.savefig("pylab1pt1_vel_time_plot.png")

#Velocity vs. position plot
plt.figure(figsize=(10,8))
plt.plot(y, v, color='k')
plt.xlabel("Position (m)", fontsize=20)
plt.ylabel("Velocity (m/s)", fontsize=20)
plt.title("Velocity vs. Position for the Spring-Mass System (Forward Euler Method)", fontsize=25)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.grid()
#plt.savefig("pylab1pt1_vel_pos_plot.png")
```

```python
#Also calculate the energy at each step using the Forward Euler method

#Define constants
dt = 0.01              #Timestep Delta t in seconds
k = 14.6               #Spring constant in N/m
m = 0.2                #Mass of bob in kg
Omega_0 = np.sqrt(k/m) #Angular frequency formula
t0 = 0                 #Initial time, in seconds
T = 0.735              #Period taken from motion sensor data, in seconds

ti = np.arange(0,10.0,dt) #Array of time values, in seconds

#Zero arrays for holding numerical solutions
y = np.zeros(len(ti)) #Array for position
v = np.zeros(len(ti)) #Array for velocity
E = np.zeros(len(ti)) #Array for energy

#Initial conditions
y[0] = 0.021875     #Amplitude of experiment in m
v[0] = 0.0          #Initial velocity before release in m/s
E[0] = (1/2)*k*y[0] #Energy of system at t = 0

#Numerical integration to solve for position, velocity, and energy
for i in range(0,len(ti)-1): 
    y[i + 1] = y[i] + (dt*v[i]) 
    v[i + 1] = v[i] - (dt*(Omega_0**2)*y[i]) 
    E[i] = ((1/2)*m*(v[i]**2)) + ((1/2)*k*(y[i]**2)) 

#Energy vs. time plot  
plt.figure(figsize=(10,8))
plt.plot(ti, E, color='k')
plt.xlabel("Time (s)", fontsize=20)
plt.ylabel("Energy (J)", fontsize=20)
plt.title('Energy vs. Time for the Spring-Mass System (Forward Euler Method)', fontsize=25)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.grid()
#plt.savefig("pylab1pt1_energy_time_plot.png")
```

```python
#Solve the same problems using the Symplectic Euler method

#Define constants
dt = 0.01              #Timestep Delta t in seconds
k = 14.6               #Spring constant in N/m
m = 0.2                #Mass of bob in kg
Omega_0 = np.sqrt(k/m) #Angular frequency formula
t0 = 0                 #Initial time, in seconds
T = 0.735              #Period taken from motion sensor data, in seconds

ti = np.arange(0,10.0,dt) #Array of time values, in seconds

#Zero arrays for holding numerical solutions
y = np.zeros(len(ti)) #Array for position
v = np.zeros(len(ti)) #Array for velocity
E = np.zeros(len(ti)) #Array for energy

#Initial conditions
y[0] = 0.021875     #Amplitude of experiment in m
v[0] = 0.0          #Initial velocity before release in m/s
E[0] = (1/2)*k*y[0] #Energy of system at t = 0

#Numerical integration to solve for position, velocity, and energy
for i in range(0,len(ti)-1): 
    y[i + 1] = y[i] + (dt*v[i]) 
    v[i + 1] = v[i] - (dt*(k/m)*y[i + 1]) 
    E[i] = ((1/2)*m*(v[i]**2)) + ((1/2)*k*(y[i]**2)) 
    
#Position vs. time plot
plt.figure(figsize=(10,8))
plt.plot(ti, y, color='k')
plt.xlabel("Time (s)", fontsize=20)
plt.ylabel("Distance (m)", fontsize=20)
plt.title("Position vs. Time for the Spring-Mass System (Symplectic Euler Method)", fontsize=25)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.grid()
#plt.savefig("pylab1pt1_pos_time_symplectic.png")

#Velocity vs. time plot
plt.figure(figsize=(10,8))
plt.plot(ti, v, color='k')
plt.xlabel("Time (s)", fontsize=20)
plt.ylabel("Velocity (m/s)", fontsize=20)
plt.title("Velocity vs. Time for the Spring-Mass System (Symplectic Euler Method)", fontsize=25)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.grid()
#plt.savefig("pylab1pt1_vel_time_symplectic.png")

#Velocity vs. position plot
plt.figure(figsize=(10,8))
plt.plot(y, v, color='k')
plt.xlabel("Position (m)", fontsize=20)
plt.ylabel("Velocity (m/s)", fontsize=20)
plt.title("Velocity vs. Position for the Spring-Mass System (Symplectic Euler Method)", fontsize=25)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.grid()
#plt.savefig("pylab1pt1_vel_pos_symplectic.png")

#Energy vs. time plot  
plt.figure(figsize=(10,8))
plt.plot(ti, E, color='k')
plt.xlabel("Time (s)", fontsize=20)
plt.ylabel("Energy (J)", fontsize=20)
plt.title("Energy vs. Time for the Spring-Mass System (Symplectic Euler Method)", fontsize=25)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.grid()
#plt.savefig("pylab1pt1_energy_time_symplectic.png")
```
