```python
#Import required packages
import numpy as np
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Load experimental data from .txt files
battery1_V, battery1_verr, battery1_A, battery1_aerr = np.loadtxt('Output_Resis_Battery1.txt', comments="#", skiprows=2, delimiter=',', unpack = True)
power1_6_5_V, power1_6_5_verr, power1_6_5_A, power1_6_5_aerr = np.loadtxt('Output_Resis_Power1_6_5.txt', comments="#", skiprows=2, delimiter=',', unpack = True)
power1_10_V, power1_10_verr, power1_10_A, power1_10_aerr = np.loadtxt('Output_Resis_Power1_10.txt', comments="#", skiprows=2, delimiter=',', unpack = True)
power1_15_V, power1_15_verr, power1_15_A, power1_15_aerr = np.loadtxt('Output_Resis_Power1_15.txt', comments="#", skiprows=2, delimiter=',', unpack = True)
power1_20_V, power1_20_verr, power1_20_A, power1_20_aerr = np.loadtxt('Output_Resis_Power1_20.txt', comments="#", skiprows=2, delimiter=',', unpack = True)
battery2_V, battery2_verr, battery2_A, battery2_aerr = np.loadtxt('Output_Resis_Battery2.txt', comments="#", skiprows=2, delimiter=',', unpack = True)
power2_6_5_V, power2_6_5_verr, power2_6_5_A, power2_6_5_aerr = np.loadtxt('Output_Resis_Power2_6_5.txt', comments="#", skiprows=2, delimiter=',', unpack = True)
power2_10_V, power2_10_verr, power2_10_A, power2_10_aerr = np.loadtxt('Output_Resis_Power2_10.txt', comments="#", skiprows=2, delimiter=',', unpack = True)
power2_15_V, power2_15_verr, power2_15_A, power2_15_aerr = np.loadtxt('Output_Resis_Power2_15.txt', comments="#", skiprows=2, delimiter=',', unpack = True)
power2_20_V, power2_20_verr, power2_20_A, power2_20_aerr = np.loadtxt('Output_Resis_Power2_20.txt', comments="#", skiprows=2, delimiter=',', unpack = True)
```

```python
#Define functions
#Write a linear model function
def model_function(x,a,b):
    return a*x + b

#Write a function to calculate the reduced chi-squared value for the fits
def red_chi_squared(y,sigma,ypred):
    a = np.zeros(len(y))
    for i in range(0,len(y)):
        a[i] = ((y[i] - ypred[i])/sigma[i])**2
    chi_squared_red = (1/(N - n))*np.sum(a)
    return chi_squared_red
```

```python
#Battery, circuit option 1

#Curve fit of the battery using circuit option 1
p_opt, p_cov = curve_fit(model_function, battery1_V, battery1_A, p0=(1,0), sigma=battery1_verr, absolute_sigma=True)
p_opt_battery1 = p_opt
p_cov_battery1 = p_cov

#Plot the data for battery, circuit option 1
plt.figure(figsize=(10,8))
plt.errorbar(battery1_V,battery1_A,xerr=battery1_verr, yerr=battery1_aerr,ls='',marker='o',lw=1, color='k', label='Data')
plt.plot(battery1_V, model_function(battery1_V,*p_opt_battery1), color='r', label='Fit')
plt.xlabel('Voltage (V)', fontsize=20)
plt.ylabel('Current (A)', fontsize=20)
plt.title('Plot of data for battery, option 1', fontsize=25)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.grid()
plt.legend()
#plt.savefig('battery1.pdf')
print("The resistance for the battery, circuit option 1 is", 1/p_opt_battery1[0], "+/-", np.abs((1/p_opt_battery1[0])*np.sqrt(p_cov_battery1[0][0])), 'Ohms.')
print()

#Calculate the reduced chi-squared value for the battery, circuit option
N = 8 #Number of data points
n = 2 #Number of parameters
print("The reduced chi-squared value for the battery, circuit option 1, is", red_chi_squared(battery1_A,battery1_aerr, model_function(battery1_V,*p_opt_battery1)),".")
```
